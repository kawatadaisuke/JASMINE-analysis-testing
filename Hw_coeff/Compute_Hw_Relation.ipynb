### Compute Hw Relation using Hw_coeff.py

Compute the definition of Hw with H and J-H using JASMINE-Mission/telescope_baseline/src/telescope_baseline/photometry/Hw_coeff.py by Kohei Miyakawa

```python
import numpy as np
from telescope_baseline.photometry import Hw_coeff
```

Compute Hw definition when the wavelength range is $10,000-16,000$ $\mathring A$ . 

```python
Hw_l = 10000.0
Hw_u = 16000.0
res, sigma, colors, ar_J_H, ar_Hw_H, residuals = Hw_coeff.compute_Hw_relation(Hw_l, Hw_u)
```

```python
Hw_coeff. plot_Hwfit(Hw_l,Hw_u,res, sigma, colors, ar_J_H, ar_Hw_H, residuals)
```

```python
# coefficient a and b
a_str = str('{:.5f}'.format(res.x[0]))
b_str = str('{:.5f}'.format(res.x[1]))
c_str =  str('{:.5f}'.format(1.0-res.x[1]))
print_text = 'Hw-H = '+a_str+' (J-H)^2$+ '+b_str+' (J-H)' 
print(print_text)
# Hw = 
print('Hw = ', b_str, ' J', ' + ', c_str, 'H', ' + ', a_str, '(J-H)^2')
```
